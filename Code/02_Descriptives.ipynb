```python
###### Decision Tree feature selection 
from sklearn.ensemble import RandomForestClassifier

feat_labels = X_in.columns[0:]

forest = RandomForestClassifier(n_estimators=500, random_state=1, class_weight="balanced")

forest.fit(X_in, y_in)

importances = forest.feature_importances_
indices = np.argsort(importances)[::-1]

for f in range(X_in.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30,
                            feat_labels[indices[f]],
                            importances[indices[f]]))
    
plt.figure(figsize=(15, 9), dpi=100)    
plt.title('Feature Importance')
plt.bar(range(X_in.shape[1]),
        importances[indices],
        align='center')
plt.xticks(range(X_in.shape[1]), feat_labels[indices], rotation=90)
plt.xlim([-1, X_in.shape[1]])
plt.savefig(f"05_output/fs_rt_balanced.pdf", bbox_inches='tight')
plt.show()
```
