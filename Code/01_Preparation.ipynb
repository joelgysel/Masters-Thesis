```python
# Import data as panda
data = pd.read_excel("03_data/data02.xlsx")
data
```

```python
# Standardization of continuous variables
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()

variables = ["Year", "Cont_tot", "Cont_area", "History", "TEDY", "TEDX", "DSER", 
             "SEDY", "STDR", "PEDY", "ResG", "OPEN", "RGRWT", "CAY", "WX", "IMF", "XG", "MG", 
             "FDIG", "INF", "M2R", "FDIY", "PR", "OVER", "MAC", "DOil", "UST", "EU"]

variables_dummies = ["Code", "Year", "Default", "Cont_tot", "Cont_area", "History", "IMF", "MAC", "DOil", "EU"]

variables_continuous = ["TEDY", "TEDX", "DSER", "SEDY", "STDR", "PEDY", "ResG", 
                        "OPEN", "RGRWT", "CAY", "WX", "XG", "MG", "FDIG",
                        "INF", "M2R", "FDIY", "PR", "OVER", "UST"]

# Generate X and Y 
dummies    = data[variables_dummies]
continuous = data[variables_continuous]
continuous = sc_x.fit_transform(continuous, )
continuous = pd.DataFrame(continuous)
continuous.columns = variables_continuous

# Create new dataframge
data_transformed = pd.concat([dummies, continuous], axis=1)
data_transformed
```

```python
pd_id = data_transformed.drop_duplicates(subset = "Code")
pd_id = pd_id[["Code"]]
pd_id.shape

import random
random.seed(1)
a = [0,1,2,3,4,5,6]*10
random.shuffle(a)
type(a)

pd_id["wookie"] = a
pd_id = pd_id[['Code', 'wookie']]

tips_summed = pd_id.groupby(["wookie"])["wookie"].count()
tips_summed
```

```python
pd_data = data_transformed.sort_values(by=["Code"], ascending=[True])
pd_id   = pd_id.sort_values(by=["Code"], ascending=[True])

pd_data = pd_data.merge(pd_id, on=["Code"], how="inner")

pd_data
```
